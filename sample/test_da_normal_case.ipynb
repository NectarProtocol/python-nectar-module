```python
#### Aggregate type: Linear Regression, Mean, Count, Variance, Minimum, Maximum, Sum
```

```python
def linear_regression_func():
    import pandas as pd
    from sklearn.linear_model import LinearRegression
    df = pd.read_csv("/app/data/worker-data.csv")
    df_filtered = df[df['age'] > 50]
    x_values = ["heart_rate","age"]
    y_values = 'height'
    model = LinearRegression() 
    X_train = df_filtered[x_values]
    Y_train = df_filtered[y_values]
    model.fit(X_train, Y_train)
    return {
        "coef": model.coef_.tolist(),
        "intercept": float(model.intercept_),
        "count": len(df_filtered),
    }
```

```python
def count_func():
    import pandas as pd
    df = pd.read_csv("/app/data/worker-data.csv")
    return {"count": len(df)}
```

```python
def mean_func():
    import pandas as pd
    df = pd.read_csv("/app/data/worker-data.csv")
    count =  len(df)
    mean = df['age'].mean()
    return {"count": count, "mean":mean}
```

```python
def variance_func():
    import pandas as pd
    df = pd.read_csv("/app/data/worker-data.csv")
    count =  len(df)
    mean = df['age'].mean()
    variance = df['age'].var()
    return {"count": count, "mean":mean,"variance":variance }
```

```python
#Single data node
```

```python
result = nectar_client.byoc_query(
    count_func, 
    bucket_ids=[bucket_ids[1]],
    policy_indexes=[0],
    use_allowlists=[True],
    access_indexes=[0]
)
```
